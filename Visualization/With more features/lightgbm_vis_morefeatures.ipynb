```python
import sys
import numpy as np
# Update sys.path to include the correct directory
sys.path.append(r"C:\\Local Disk (A)\\Github Files\\Projects\\Mass-Sceptra-Classification\\Visualization\\With more features")

# Paths to the datasets
dataset_paths = [
    r"C:\\Local Disk (A)\\Github Files\\Projects\\Mass-Sceptra-Classification\\Datasets\\CM1",
    r"C:\\Local Disk (A)\\Github Files\\Projects\\Mass-Sceptra-Classification\\Datasets\\CM2",
    r"C:\\Local Disk (A)\\Github Files\\Projects\\Mass-Sceptra-Classification\\Datasets\\CM3"
]

def combine_datasets(dataset_paths):
    """Combines multiple datasets into one."""
    X_combined, y_combined = [], []

    for path in dataset_paths:
        print(f"Processing dataset at {path}...")
        X, y, _ = gf.generate_balanced_comparisons(path, debug=True)  # Generate features and labels
        if X is not None:
            X_combined.append(X)
            y_combined.append(y)

    # Combine all datasets
    if X_combined and y_combined:
        X_combined = np.vstack(X_combined)
        y_combined = np.concatenate(y_combined)
        return X_combined, y_combined
    else:
        print("No data generated from the datasets. Exiting...")
        return None, None

def main():
    # Combine datasets
    print("Combining datasets...")
    X, y = combine_datasets(dataset_paths)
    if X is None:
        return

    # Split into train/test sets and scale features
    print("Splitting and scaling the combined dataset...")
    X_train, X_test, y_train, y_test, scaler = gf.prepare_dataset(dataset_paths[0])  # Dummy call for scaler
    X_train, X_test, y_train, y_test = gf.train_test_split(
        X, y, test_size=0.4, random_state=42, stratify=y
    )
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    # Initialize the LightGBM model
    print("Initializing the LightGBM model...")
    model = gf.initialize_lightgbm_model()

    # Perform cross-validation
    print("Performing cross-validation...")
    cv_scores = gf.perform_cross_validation(model, X_train, y_train)

    # Train and evaluate the model
    print("Training and evaluating the model...")
    y_pred, y_prob = gf.train_and_evaluate(model, X_train, y_train, X_test, y_test)

    # Plot feature importance
    print("Plotting feature importance...")
    gf.plot_feature_importance(model, feature_names=['Cosine Similarity', 'Area Ratio', 'Std Ratio'])

    # Plot confusion matrix
    print("Plotting confusion matrix...")
    gf.plot_confusion_matrix(y_test, y_pred)

    # Plot correlation
    print("Plotting correlation...")
    gf.plot_correlation(y_test, y_prob)

if __name__ == "__main__":
    main()
```

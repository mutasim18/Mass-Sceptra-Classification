```python
import sys
import numpy as np
sys.path.append(r"C:\\Local Disk (A)\\Github Files\\Projects\\Mass-Sceptra-Classification\\Visualization")
import global_fun as gf  # Import the custom functions

# Paths to the datasets
dataset_paths = [
    r"C:\\Local Disk (A)\\Github Files\\Projects\\Mass-Sceptra-Classification\\Datasets\\CM1",
    r"C:\\Local Disk (A)\\Github Files\\Projects\\Mass-Sceptra-Classification\\Datasets\\CM2",
    r"C:\\Local Disk (A)\\Github Files\\Projects\\Mass-Sceptra-Classification\\Datasets\\CM3"
]

def combine_datasets(dataset_paths):
    """Combines multiple datasets into one."""
    X_combined, y_combined = [], []

    for path in dataset_paths:
        print(f"Processing dataset at {path}...")
        X, y, _ = gf.generate_balanced_comparisons(path, debug=False)  # Generate features and labels
        if X is not None:
            X_combined.append(X)
            y_combined.append(y)

    # Combine all datasets
    if X_combined and y_combined:
        X_combined = np.vstack(X_combined)
        y_combined = np.concatenate(y_combined)
        return X_combined, y_combined
    else:
        print("No data generated from the datasets. Exiting...")
        return None, None

def main():
    # Combine datasets
    print("Combining datasets...")
    X_combined, y_combined = combine_datasets(dataset_paths)
    if X_combined is None or y_combined is None:
        print("Dataset combination failed. Exiting...")
        return

    # Split combined dataset into training and testing sets
    X_train, X_test, y_train, y_test = gf.train_test_split(
        X_combined, y_combined, test_size=0.4, random_state=42, stratify=y_combined
    )

    # Initialize the Random Forest model
    print("Initializing the Random Forest model...")
    model = gf.initialize_random_forest_model()

    # Perform cross-validation
    print("Performing cross-validation...")
    cv_scores = gf.perform_cross_validation(model, X_train, y_train)

    # Train and evaluate the model
    print("Training and evaluating the model...")
    y_pred, y_prob = gf.train_and_evaluate(model, X_train, y_train, X_test, y_test)

    # Plot feature importance
    print("Plotting feature importance...")
    gf.plot_feature_importance(model, feature_names=['Cosine Similarity'])

    # Plot correlation between true labels and predicted probabilities
    print("Plotting correlation...")
    gf.plot_correlation(y_test, y_prob)

    # Plot confusion matrix
    print("Plotting confusion matrix...")
    gf.plot_confusion_matrix(y_test, y_pred)

if __name__ == "__main__":
    main()
```

```python
# bar graph of all the performances
import matplotlib.pyplot as plt
import numpy as np

def plot_stats(stats, datasets, metrics):
    """
    Creates a bar plot to showcase statistics for multiple datasets.

    :param stats: A list of dictionaries containing stats for each dataset.
                  Each dictionary should have keys matching the metrics.
    :param datasets: A list of dataset names (e.g., ['CM1', 'CM2', 'CM3', 'Combined']).
    :param metrics: A list of metric names to be plotted (e.g., ['Accuracy', 'ROC AUC', 'Cosine Similarity', 'Correlation']).
    """
    num_datasets = len(datasets)
    num_metrics = len(metrics)
    
    # Set the positions and width for the bars
    x = np.arange(num_datasets)
    width = 0.15

    # Create the bar plot
    plt.figure(figsize=(20, 10))
    for i, metric in enumerate(metrics):
        metric_values = [dataset_stats.get(metric, 0) for dataset_stats in stats]
        plt.bar(x + i * width, metric_values, width, label=metric)

    # Add labels, title, and legend
    plt.xlabel('Datasets', fontsize=12)
    plt.ylabel('Scores', fontsize=12)
    plt.title('Performance Metrics for Datasets', fontsize=14)
    plt.xticks(x + width * (num_metrics - 1) / 2, datasets, fontsize=12)
    plt.legend(title="Metrics", fontsize=10)
    plt.tight_layout()
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Show the plot
    plt.show()

# Example stats for CM1, CM2, CM3, and Combined datasets
stats = [
    {'Accuracy': 0.9517, 'ROC AUC': 0.9742,  'Correlation': 0.8716},  # CM1
    {'Accuracy': 0.9626, 'ROC AUC': 0.9784,  'Correlation': 0.9093},  # CM2
    {'Accuracy': 0.9938, 'ROC AUC': 0.9999,  'Correlation': 0.9824},  # CM3
    {'Accuracy': 0.9496, 'ROC AUC': 0.9765,  'Correlation': 0.8671}   # Combined
]

# Dataset names
datasets = ['CM1', 'CM2', 'CM3', 'Combined']

# Metrics to display
metrics = ['Accuracy', 'ROC AUC', 'Correlation']

# Call the function to create the bar plot
plot_stats(stats, datasets, metrics)
```

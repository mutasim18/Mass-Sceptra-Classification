```python
import sys
sys.path.append(r"C:\\Local Disk (A)\\Github Files\\Projects\\Mass-Sceptra-Classification\\Visualization")
import global_fun as gf  # Import the custom functions

# Path to the CM1 dataset
dataset_path = r"C:\\Local Disk (A)\\Github Files\\Projects\\Mass-Sceptra-Classification\\Datasets\\CM1"

def main():
    # Prepare the dataset
    print(f"Processing dataset at {dataset_path}...")
    X_train, X_test, y_train, y_test, scaler = gf.prepare_dataset(dataset_path)
    if X_train is None:
        print("Dataset preparation failed. Exiting...")
        return

    # Initialize the Random Forest model
    print("Initializing the Random Forest model...")
    model = gf.initialize_random_forest_model()

    # Perform cross-validation
    print("Performing cross-validation...")
    cv_scores = gf.perform_cross_validation(model, X_train, y_train)

    # Train and evaluate the model
    print("Training and evaluating the model...")
    y_pred, y_prob = gf.train_and_evaluate(model, X_train, y_train, X_test, y_test)

    # Plot feature importance
    print("Plotting feature importance...")
    gf.plot_feature_importance(model, feature_names=['Cosine Similarity'])

    # Plot correlation between true labels and predicted probabilities
    print("Plotting correlation...")
    gf.plot_correlation(y_test, y_prob)

    # Plot confusion matrix
    print("Plotting confusion matrix...")
    gf.plot_confusion_matrix(y_test, y_pred)

if __name__ == "__main__":
    main()
```
